# Simulated Annealing used to optimize Assemble to Order Problem

Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import imageio
import os

from sa4ato import discrete_stochastic_simulated_annealing as sa
from sa4ato.assemble_to_order import ato
from sa4ato.assemble_to_order import ATOSystem
```

## Assemble to Order System

### Moderately Problem instance

```python
print(f'Solution space: {21**8}')
```

Set seeds

```python
np.random.seed(0)
random.seed(0)
```

### Assemble to order model function

    :param inventory_level: the starting inventory level as well as the inventory level goal
    :param t_warm_up: periods without tracked statistics
    :param t_max: periods
    :param ato_system: specifies the size of the ato system
    :return: negative mean profit over t_max-t_warm_up periods

```python
inventory_level = np.array([3, 4, 3, 7, 5, 6, 4, 3])
profit = -ato(inventory_level=inventory_level, t_warm_up=20, t_max=70, ato_system=ATOSystem.MODERATELY)
profit2 = -ato(inventory_level=inventory_level, t_warm_up=20, t_max=70, ato_system=ATOSystem.MODERATELY)
print(f'because of the stochastic background of the assemble to order problem the profit can vary: {profit}, {profit2}')
```

## Simulated Annealing used so solve the Assemble to Order Problem

```python
#np.random.seed(0)
x0 = np.random.randint(0, 21, 8)

#steps per SA
step_max=2500

# plot parameters
figsize = (6,4)
dpi = 200
```

```python
opt_count_direct_2 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=2, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_count_direct_1 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_count_direct_01 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=0.1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_count_direct_001 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=0.01, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)

print('1/4 FINISHED')

opt_count_random_2  = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=2, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_count_random_1 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_count_random_01 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=0.1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_count_random_001 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.COUNT,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=0.01, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)

print('2/4 FINISHED')

opt_mean_direct_2 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=2, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_mean_direct_1 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_mean_direct_01 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=0.1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_mean_direct_001 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.DIRECT_NEIGHBOURHOOD, t=0.01, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)

print('3/4 FINISHED')

opt_mean_random_2 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=2, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_mean_random_1 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_mean_random_01 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=0.1, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)
opt_mean_random_001 = sa.Minimize(ato, x0, opt_solution=sa.OptSolution.MEAN,
                  neighbourhood_func=sa.NeighbourhoodFunc.RANDOM_NEIGHBOURHOOD, t=0.01, bounds=[0, 20],
                  step_max=step_max, ato_system=ATOSystem.MODERATELY)

print('4/4 FINISHED')
```

### Animated inventory levels

```python
n_frames = 1
bg_color = '#FFFFFF'
bar_color = '#1f77b4'
gif_name = 'images/bars'
x = [1, 2, 3, 4, 5, 6, 7, 8]
coordinates_lists = []
for i in range(len(opt_mean_direct_01.hist)):
    if i % 10 == 0:
        coordinates_lists.append(opt_mean_direct_01.hist[i][3])
filenames = []

for index in np.arange(0, len(coordinates_lists) - 1):
    y = coordinates_lists[index]
    y1 = coordinates_lists[index + 1]
    y_path = np.array(y1) - np.array(y)

    for i in np.arange(0, n_frames + 1):
        y_temp = (y + (y_path / n_frames) * i)
        # plot
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.set_facecolor(bg_color)
        textstr = 'Iteration: ' + str((index+1)*10) + '\nProfit/Periode: ' + str(-1 * int(opt_mean_direct_01.hist[index*10][2]))
        props = dict(facecolor='white')
        plt.bar(x, y_temp, width=0.4, color=bar_color)
        plt.ylim(0, 35)
        ax.text(0.95, 0.95, textstr, transform=ax.transAxes, fontsize=11, verticalalignment='top', ha='right', bbox=props)
        #plt.legend()
        # remove spines
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        # grid
        ax.set_axisbelow(True)
        ax.yaxis.grid(color='gray', linestyle='dashed', alpha=0.7)
        # build file name and append to list of file names
        filename = f'images/frame_{index}_{i}.png'
        filenames.append(filename)

        # last frame of each viz stays longer
        if index == len(coordinates_lists) - 2:
            for j in range(10):
                filenames.append(filename)
        # save img
        plt.savefig(filename, dpi=96, facecolor=bg_color)
        plt.close()

# Build GIF
with imageio.get_writer(f'{gif_name}.gif', mode='I', fps=60) as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
# Remove files
for filename in set(filenames):
    os.remove(filename)
```
